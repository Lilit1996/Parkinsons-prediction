### Exploratory Data Analysis

```python
#Importing libraries
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_palette('YlOrRd_r', n_colors = 3)

import warnings
warnings.filterwarnings('ignore')
```

```python
#Reading and printing data
df = pd.read_csv("../data/parkinsons_data.txt")
df.head(10)
```

```python
df.info()
```

```python
#Checking missing values
df.isna().sum()
```

```python
#Checking number of unique values
df.nunique()
```

```python
df.describe().transpose()
```

```python
#Finding out target variable distribution
df["status"].value_counts(normalize=True)
```

The code above shows that the target variable is imbalanced with ~ 75% with Parkinson's disease and ~ 25% healthy.

```python
sns.countplot(df["status"].values)
plt.xlabel('Parkinson Values')
plt.ylabel('Counts')
plt.show()
```

Base model : all have PD -> 0.754 accuracy
In the next stages the accuracy of the model predictions will be benchmarked against base model of 0.754.

```python
fig, axes = plt.subplots(22,1, figsize=(10, 80))

for i, col in enumerate(df.columns.difference(["status", "name"])):
    disease = df[df["status"] == 1][col].values
    healty = df[df["status"] == 0][col].values
    sns.distplot(disease, ax=axes[i], hist=True, label="Parkinson's Disease Cases")
    sns.distplot(healty, ax=axes[i], hist=True, label="Healthy Cases")
    axes[i].set_title(f"{col} Distribution plot")
    axes[i].legend()
```

```python
#Plotting the boxplots of the measures and seeing the outliers
for i, col in enumerate(df.columns.difference(["status", "name"])):
    disease = df[df["status"] == 1][col].values
    healty = df[df["status"] == 0][col].values
    plt.boxplot([disease, healty])
    plt.title(f"{col} Box plot")
    plt.xticks([1, 2], ["Parkinson's Disease Cases", "Healthy Cases"])
    plt.show()
```

After checking the outliers, we decided to leave them in the data, as the number of instances is not large.

```python
#Printing the correlation between the features and the target variable
corr_values=df.corr()['status']
corr_values.abs().sort_values(ascending=False)
```

```python
#Plotting the correlation heatmap
corr = df.corr()
plt.figure(figsize=(15,8))
ax = sns.heatmap(corr, annot=True)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
plt.show()
```

The heatmap shows that correlation beetween MDVP:Jitter(%) and MDVP:Jitter(Abs), MDVP:RAP and MDVP:Jitter(%), MDVP:PPQ and MDVP:Jitter(%), MDVP:PPQ and MDVP:RAP, Jitter:DDP and MDVP:Jitter(%),  Jitter:DDP and MDVP:PPQ, Jitter:DDP and MDVP:RAP, MDVP:Shimmer(dB) and MDVP:Shimmer, Shimmer:APQ3 and MDVP:Shimmer, Shimmer:APQ3 and MDVP:Shimmer(dB),  Shimmer:APQ5 and MDVP:Shimmer, Shimmer:APQ5 and MDVP:Shimmer(dB), Shimmer:APQ5 and Shimmer:APQ3, Shimmer:DDA and MDVP:Shimmer,  Shimmer:DDA and MDVP:Shimmer(dB), Shimmer:DDA and Shimmer:APQ3, Shimmer:DDA and Shimmer:APQ5, Shimmer:DDA and MDVP:APQ, PPE and spread1 are high.
